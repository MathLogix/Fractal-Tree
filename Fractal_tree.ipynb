```python
# Random fractal tree with Matplotlib

import matplotlib.pyplot as plt
import numpy as np
import random

def fractal_tree(ax, x, y, angle, branch_length, level, pensize):
    if level == 0:
        return
    
    x_end = x + branch_length * np.cos(np.radians(angle))
    y_end = y + branch_length * np.sin(np.radians(angle))
    
    ax.plot([x, x_end], [y, y_end], color="white", linewidth=pensize)
    
    num_branches = random.randint(2, 4)
    
    for _ in range(num_branches):
        new_angle = angle + random.uniform(-60, 60)  
        new_branch_length = branch_length * random.uniform(0.5, 0.8)  
        new_pensize = max(pensize - random.randint(2, 4), 1)  
        
        fractal_tree(ax, x_end, y_end, new_angle, new_branch_length, level - 1, new_pensize)

fig, ax = plt.subplots(figsize=(10, 10), facecolor="black")
ax.axis("off")
ax.set_aspect("equal")

fractal_tree(ax, 0, -300, 90, 50, 8, 10)

plt.show()
```

```python
# Regular fractal tree with Pygame

import pygame
import math

def fractal_tree(surface, x, y, angle, branch_length, level, pensize):
    if level == 0:
        return
    
    x_end = x + branch_length * math.cos(math.radians(angle))
    y_end = y - branch_length * math.sin(math.radians(angle))
    
    pygame.draw.line(surface, (255, 255, 255), (x, y), (x_end, y_end), pensize)
    
    fractal_tree(surface, x_end, y_end, angle - 30, branch_length * 0.7, level - 1, max(pensize - 3, 1))
    fractal_tree(surface, x_end, y_end, angle + 30, branch_length * 0.7, level - 1, max(pensize - 3, 1))

pygame.init()
screen_width, screen_height = 1500, 800
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Fractal Tree")
screen.fill((0, 0, 0))

start_x, start_y = screen_width // 2, screen_height - 50 
fractal_tree(screen, start_x, start_y, 90, 200, 10, 20)

pygame.display.flip()

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
# Random tree with turtle

import turtle
import random

def tree(branch_len, t):
    if branch_len > 5:
        t.forward(branch_len)
        angle = random.randint(15, 45)
        length = random.randint(10, 20)
        t.right(angle)
        tree(branch_len - length, t)
        t.left(angle * 2)
        tree(branch_len - length, t)
        t.right(angle)
        t.backward(branch_len)

def main():
    t = turtle.Turtle()
    t.speed(0)
    my_win = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    tree(100, t)
    my_win.exitonclick()

main()
```

```python
# Sierpinski Triangle

import turtle

def draw_triangle(points, color, t):
    t.fillcolor(color)
    t.up()
    t.goto(points[0][0], points[0][1])
    t.down()
    t.begin_fill()
    t.goto(points[1][0], points[1][1])
    t.goto(points[2][0], points[2][1])
    t.goto(points[0][0], points[0][1])
    t.end_fill()

def midpoint(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)

def sierpinski(points, degree, t):
    colormap = ['blue', 'red', 'green', 'white', 'yellow',
                'violet', 'orange']
    draw_triangle(points, colormap[degree], t)
    if degree > 0:
        sierpinski([points[0],
                    midpoint(points[0], points[1]),
                    midpoint(points[0], points[2])],
                   degree-1, t)
        sierpinski([points[1],
                    midpoint(points[0], points[1]),
                    midpoint(points[1], points[2])],
                   degree-1, t)
        sierpinski([points[2],
                    midpoint(points[2], points[1]),
                    midpoint(points[0], points[2])],
                   degree-1, t)

def main():
    t = turtle.Turtle()
    my_win = turtle.Screen()
    my_points = [[-100, -50], [0, 100], [100, -50]]
    sierpinski(my_points, 3, t)
    my_win.exitonclick()

main()
```
